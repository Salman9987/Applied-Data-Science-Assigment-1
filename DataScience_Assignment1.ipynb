```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import csv

data = pd.read_csv('stress_detection.csv')
```

```python
mean_openness = data['Openness'].mean()
data['Openness'] = data['Openness'].fillna(mean_openness)

mean_conscientiousness = data['Conscientiousness'].mean()
data['Conscientiousness'] = data['Conscientiousness'].fillna(mean_conscientiousness)

median_sleep_time = data['sleep_time'].median()
data['sleep_time'] = data['sleep_time'].fillna(median_sleep_time)

median_wake_time = data['wake_time'].median()
data['wake_time'] = data['wake_time'].fillna(median_wake_time)

mean_extraversion = data['Extraversion'].mean()
data['Extraversion'] = data['Extraversion'].fillna(mean_extraversion)

mean_agreeableness = data['Agreeableness'].mean()
data['Agreeableness'] = data['Agreeableness'].fillna(mean_agreeableness)

mean_neuroticism = data['Neuroticism'].mean()
data['Neuroticism'] = data['Neuroticism'].fillna(mean_neuroticism)

median_sleep_duration = data['sleep_duration'].median()
data['sleep_duration'] = data['sleep_duration'].fillna(median_sleep_duration)

median_screen_on_time = data['screen_on_time'].median()
data['screen_on_time'] = data['screen_on_time'].fillna(median_screen_on_time)

median_skin_conductance = data['skin_conductance'].median()
data['skin_conductance'] = data['skin_conductance'].fillna(median_skin_conductance)

median_accelerometer = data['accelerometer'].median()
data['accelerometer'] = data['accelerometer'].fillna(median_accelerometer)

median_mobility_radius = data['mobility_radius'].median()
data['mobility_radius'] = data['mobility_radius'].fillna(median_mobility_radius)

print (data)
```
